```python
import cv2
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
def show_images(lst):
    while True:
        for i in lst:
            cv2.imshow(i[0],i[1])
        ret = cv2.waitKey(0)
        if ret==27:
            break
    cv2.destroyAllWindows()
```
